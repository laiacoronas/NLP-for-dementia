```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('C:\\Users\\lclai\\Desktop\\datamimic\\patients_dementia.csv')
data
```

```python
data.columns
```

```python
df = data[data['DEMENTIA']==1]
df = df.drop(columns=["SUBJECT_ID", "HADM_ID"], errors='ignore')
print("InformaciÃ³n del dataset:")
print(df.info())
print("\nEstadÃ­sticas descriptivas:")
print(df.describe())
```

```python
numerical_cols = ['AGE', 'LAST_ADMISSION_LENGTH']
df[numerical_cols].hist(bins=30, figsize=(10, 5))
plt.suptitle("DistribuciÃ³n de variables numÃ©ricas")
plt.show()
```

```python
categorical_cols = ['GENDER', 'ETHNICITY']
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=col, data=df, order=df[col].value_counts().index)
    plt.title(f"DistribuciÃ³n de {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
df.columns
```

```python
disease_cols = ['HYPERTENSION', 'ATRIAL_FIBRILLATION',
       'KIDNEY_FAILURE', 'UTI', 'HEART_FAILURE', 'RESPIRATORY_FAILURE',
       'PNEUMONITIS', 'DIABETES', 'CORONARY_ATHEROSCLEROSIS', 'HYPOTHYROIDISM',
       'ANEMIA', 'SEPSIS', 'HYPERLIPIDEMIA', 'PNEUMONIA', 'HYPEROSMOLALITY',
       'SEVERE_SEPSIS', 'HYPERCHOLESTEROLEMIA', 'ESOPHAGEAL_REFLUX',
       'KIDNEY_DISEASE', 'ACIDOSIS', 'HYPERTENSION_KIDNEY']
disease_counts = df[disease_cols].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=disease_counts.index, y=disease_counts.values)
plt.xticks(rotation=90)
plt.title("Frecuencia de enfermedades")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='ETHNICITY', y='AGE', hue='DEMENTIA', data=df)
plt.xticks(rotation=45)
plt.title("DistribuciÃ³n de la edad por etnicidad y diagnÃ³stico de demencia")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr = df[disease_cols].corr()
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title("CorrelaciÃ³n entre enfermedades")
plt.show()
```
